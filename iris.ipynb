```python
x = df1.iloc[:,0:5]
y = df1.iloc[:,-1:]
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df1['Species']=le.fit_transform(df1['Species'])
df1
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=143)
x_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
classfi = KNeighborsClassifier(n_neighbors=5,p=2)
classfi.fit(x_train,y_train)
```

```python
y_predict = classfi.predict(x_test)
y_predict
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(y_predict,y_test)
print("confusion_matrix",cm)
accscore = accuracy_score(y_predict,y_test)
print("accuracy_score",accscore)
```

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(x_train,y_train)
```

```python
y_predict = svc.predict(x_test)
y_predict
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(y_predict,y_test)
print("confusion_matrix",cm)
accscore = accuracy_score(y_predict,y_test)
print("accuracy_score",accscore)
```

```python
from sklearn.linear_model import LogisticRegression
Log = LogisticRegression()
Log.fit(x_train,y_train)
```

```python
y_pred = Log.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(y_pred,y_test)
print("confusion_matrix",cm)
accscore = accuracy_score(y_pred,y_test)
print("accuracy_score",accscore)
```
